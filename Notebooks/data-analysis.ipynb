We can almost see that as temperature increases along the area, generally, the ozone quality is higher regardless of whether the area is in a decline (besides in Florida and Washington which have low air pollution). We can almost see a slight curve upwards from the bottom left toward Arizona at the right, with California presenting itself as a clear outlier in terms of how bad the air quality is. Generally, heat and temperature across all areas can be seen as correlated with ozone, whether that be by the line graph or by the area scatter plot.

### Commuting In 2020
------------------------------------------------------------------------------------------------------------------
With the drop in commuting due to the quarantine, commuting data has been tracked and visualizing and understanding this data in comparison to particulate matter may reveal some interesting correlations.

Let's look at commuting in Detroit during the quarantine.

```python
mobility_plot(mi_mobility, 'Detroit Metro Commuting 2020')
correlation(mi_mobility, 'Detroit Metro Commuting 2020')
```

Detroit seems to have a strong correlation between particulate matter and vehicle traffic. Although it seems ozone also has a strong correlation, the numbers simply line up due to temperature increases of the summer as the x axis of time moves closer to July. Particulate matter, as we can see with the temperature graphs and energy, does not change or spike based on coal or temperature changes. Regardless, the particulate matter correlation is higher than the ozone, especially with workplace commuting which most likely makes up a significant amount of vehicle traffic. Detroit also has the strongest correlation between particulate matter and commuting within the Central Region, indicating it's motorized vehicle production may play a role in producing PM2.5.

Let's take a look at the Northeast, specifically the New York Metro and Boston Metro.

```python
mobility_plot(ny_nj_pa_mobility, 'New York Metro Commuting 2020')
mobility_plot(ma_nh_mobility, 'Greater Boston Metro Commuting 2020')
```

We can see that both the New York Metro and Boston Metro's particulate matter trends fall and rise alongside the commuting trend. The New York Metro trend seems to follow more closely alongside commuting than Boston's fluctuations, as New York traffic is most likely greater than the Boston Metro. They both have relatively strong correlations between commuting and particulate matter.

Let's take a look at the Northeast commuting correlation matrix.

```python
correlation(northeast_mobility, 'Northeast Commuting 2020')
```

We can see that particulate matter has a strong correlation across the entire Northeast area. The MSAs of this region all have high densities: within a population density ranking of these areas, we find the four areas of the northeast all rest within the top 6 in density levels. 

The ozone in all of these areas also has a very weak correlation, supporting the idea that ozone is more caused by temperature rather than motorized vehicles. 

The two other areas within the top 6 areas of density are the Los Angeles Metro and the Minneapolis-Saint Paul Metro. Let's take a look at them.

```python
mobility_plot(mn_wi_mobility, 'Minneapolis-Saint Paul Metro Commuting 2020')
mobility_plot(ca_mobility, 'Los Angeles Metro Commuting 2020')
```

Both Minneapolis and Los Angeles have subtle correlations between particulate matter and commuting. However, one thing to note is that among the Southwest Region that Los Angeles is a part of, only this MSA has any correlation between commuting and particulate matter trends. 

Los Angeles may be infamously known for it's "LA Traffic," while the other regions in Texas and Arizona are more rural and less urbanized areas that have low population densities.
