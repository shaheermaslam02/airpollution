### Import Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from glob import glob
```

### Helpful Methods
Here we create some methods to make graphing the geographic areas much easier!

```python
def concatinate(y):
    
    path = y
    data = glob(path + '*csv')
    data.sort()
    return pd.concat([pd.read_csv(f) for f in data])

def createDataframe(x):
    x['ozone_rolling_mean_365'] = x['Ozone_AQI_Value'].rolling(window=365).mean()
    x['particle_rolling_mean_365'] = x['PM2.5_AQI_Value'].rolling(window=365).mean()
    data = pd.DataFrame({'X' : x['ozone_rolling_mean_365'],
                         'Y' : x['particle_rolling_mean_365'],
                         'T' : x['Date']})
    return data

def ozone_plot(x): 
    ax = x.plot.line('T','X', figsize = (15, 10), title = 'Ozone Trend')
    ax.set_xlabel('Time (1985-2020)')
    ax.set_ylabel('Air Quality Index (AQI)')
    
def particle_plot(x):
    ax = x.plot.line('T', 'Y', figsize = (15, 10), title = 'Particulate Matter Trend')
    ax.set_xlabel('Time (2000-2020)')
    ax.set_ylabel('Air Quality Index (AQI)')
    
```

### Collecting Data
We then initialize the geographic areas using the methods created earlier by accessing their data files so we can graph the air quality trends of the two main pollutants: Ozone (O3), and Particulate Matter (PM2.5).

```python
ny_nj_pa_files = concatinate('/Users/smaslam/Desktop/airpollution/airpollution/NewYork-Newark-JerseyCity/')
ny_nj_pa = (createDataframe(ny_nj_pa_files))
il_in_wi_files = concatinate('/Users/smaslam/Desktop/airpollution/airpollution/Chicago-Naperville-Elgin/')
il_in_wi = (createDataframe(il_in_wi_files))
```

### Graphing The Data
Now it's time to graph the data! Below are the core geographical areas which you can view the graphs of for both ozone trends and particulate matter trends:

### NY-NJ-PA

```python
ozone_plot(ny_nj_pa)
particle_plot(ny_nj_pa)
```
